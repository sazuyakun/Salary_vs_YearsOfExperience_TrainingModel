```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Salary_Data.csv')
yearOfExperience = data.iloc[:, :-1].values # This method makes sure you return a 2D array, using only [:, -1] will 
                                            # return a 1D array and would need to be reshaped later for training
                                            # Independent variable
salary = data.iloc[:, -1].values            # Dependent variable
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(yearOfExperience, salary, test_size=0.2, random_state=0)
                                                # x-------^    y--------^
```

#### This was the basic importing { libraries -> extracting data -> split to train }
---

## Training Linear reg on training model

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)          # Thing to remember is that the .fit() "method" is used to train our model
# Independent----^        ^------Dependent

# Make sure to pass a 2D array when passing x_train in .fit()
```

## Predicting the Test set

```python
y_pred = regressor.predict(x_test)
```

---
---

## Visualising


```python
# Training set

plt.scatter(x_train, y_train, c='r')
plt.plot(x_train, regressor.predict(x_train))
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Test set

plt.scatter(x_test, y_test, c='r')
plt.plot(x_test, y_pred)
plt.title('Salary vs Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
